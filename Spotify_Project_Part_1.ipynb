```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify = pd.read_csv("/content/drive/MyDrive/Spotify_final_dataset.csv")
```

```python
spotify
```

```python
spotify.rename(columns={"Days": "Days Since Released"}, inplace=True)
spotify['Top 10 (xTimes)'] = spotify['Top 10 (xTimes)'].astype(int)
```

```python
spotify
```

```python
spotify.isnull().sum()
```

```python
spotify.duplicated()
```

```python
spotify["Total Streams"].hist(bins = 30, color="black")
plt.title("Distribution of Total Streams")
plt.xlabel("Streams (In Hundred Millons)")
plt.ylabel("Frequency")
```

```python
spotify['Top 10 (xTimes)'].hist(bins = 50, color = 'green')
plt.title("Distribution of Songs Reaching The Top 10")
plt.xlabel("Number of Times")
plt.ylabel("Frequency")
```

```python
spotify['Peak Position'].hist(bins = 25)
plt.title("Distribution Of Peak Positions")
plt.xlabel("Top Positions")
plt.ylabel("Frequency")
```

```python
spotify["Days Since Released"].hist(bins=30, color='#1ED761')
plt.title("Distribution Of Days Since Release")
plt.xlabel("Number of Days Since Debut")
plt.ylabel("Frequency")
```

```python
sns.relplot(x="Days Since Released", y="Total Streams", data=spotify, color='green')
plt.title("Number Of Days Since Release vs Total Streams (In Hundred Millions)")
```

```python
sns.relplot(x="Top 10 (xTimes)", y="Total Streams", data=spotify)
plt.title("Times the Song Got In Top 10 Vs Total Streams")
plt.ylabel("Total Streams (In Hundred Millions)")
```
