```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


```

```python
spotify = pd.read_csv("/content/spotify_cleaned.csv")
spotify
```

```python
sns.scatterplot(data=spotify, x="Position", y="Peak Streams", color="green")
plt.title("Playlist Position vs Peak Streams")
plt.show()
```

```python
spotify["Position Bin"] = pd.cut(
    spotify["Position"],
    bins=[0, 100, 500, 1000, 2500, 5000, spotify["Position"].max()],
    labels=["1–100", "101–500", "501–1000", "1001–2500", "2501–5000", "5000+"]
)

plt.figure(figsize=(10, 6))
sns.boxplot(data=spotify, x="Position Bin", y="Peak Streams", color="green")
plt.title("Position Tier vs Peak Streams")
plt.xlabel("Position Range")
plt.ylabel("Peak Streams")
plt.tight_layout()
plt.show()
```
